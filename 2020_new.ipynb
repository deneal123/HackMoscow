```python
pd.set_option('display.max_colwidth', None)

def process_substring(value, substring):
    if substring in value:
        new_value = value.split(substring, 1)[1].strip()
        return new_value
    return value

substrings_to_process = [
    'зд. ',
    'зд.',
    'здание',
    'по тарифам  ',
    'по тарифам ',
    'по тарифам,',
    'ул.',
    'ул. ',
    'г.',
    'с.',
    
]

for substring in substrings_to_process:
    df['Направление расходов'] = df['Направление расходов'].apply(lambda x: process_substring(x, substring))

def remove_suffix_from_column(column, suffix):
    # Создаем новый список для хранения строк без суффикса
    new_column = []

    # Перебираем каждую строку в столбце
    for row in column:
        # Проверяем, заканчивается ли текущая строка на заданный суффикс
        if row.endswith(suffix):
            # Если да, удаляем суффикс и добавляем строку в новый список
            new_row = row[:-len(suffix)]
            new_column.append(new_row)
        else:
            # Если нет, просто добавляем строку в новый список
            new_column.append(row)

    return new_column

suffixes_to_remove = [
    "-  ст.134.1100",
    " ст.134.1100",
    "-",
    " -",
    ",",
    "- ",
    "- ст. 134.1100",
    " -  ",
    ", "
    "Здание ",
    " -ст.134.1100",
    ". Здание ",
    "-ст.134.1100",
    " ст. 134.1100",
    "-ст.1341100",
    "ст.134.1100"
]

for suffix in suffixes_to_remove:
    df['Направление расходов'] = remove_suffix_from_column(df['Направление расходов'], suffix)
    
# Выведите уникальные значения
for value in df['Направление расходов'].unique():
    print(value)
```

```python
def replace_column(dataframe, column_name, new_values):
    dataframe[column_name] = new_values
    return dataframe

df = replace_column(df, 'Наименование вида расходов', type_JKY)

df.head()
```

```python
def swap_columns(dataframe, column1, column2):
    dataframe[column1], dataframe[column2] = dataframe[column2].copy(), dataframe[column1].copy()
    dataframe.rename(columns={column1: column2, column2: column1}, inplace=True)
    return dataframe

df = swap_columns(df, 'Направление расходов', 'Наименование вида расходов')

df.head()
```
