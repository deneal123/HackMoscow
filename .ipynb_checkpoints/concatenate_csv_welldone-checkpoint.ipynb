```python
#Импортируем библиотеки
import numpy as np
import pandas as pd
import copy
import re
import openpyxl
```

```python
# Читаем таблицу 2019.
df_2019 = pd.read_csv("C:/Users/NightMare/Desktop/YandexHack/DATA_well_done/2019_welldone.csv", encoding='Windows-1251')
# Читаем таблицу 2020_old.
df_2020_old = pd.read_csv("C:/Users/NightMare/Desktop/YandexHack/DATA_well_done/2020_old_welldone.csv", encoding='Windows-1251')
# Читаем таблицу 2020_new.
df_2020_new = pd.read_csv("C:/Users/NightMare/Desktop/YandexHack/DATA_well_done/2020_new_welldone.csv", encoding='Windows-1251')
# Читаем таблицу 2021.
df_2021 = pd.read_csv("C:/Users/NightMare/Desktop/YandexHack/DATA_well_done/2021_welldone.csv", encoding='Windows-1251')
```

```python
# Список имен DataFrame, которые вы хотите объединить
df_names = [df_2019, df_2020_old, df_2020_new, df_2021]
```

```python
# Создайте пустой DataFrame, в который будут добавлены данные
combined_data = pd.DataFrame()
```

```python
# Объедините файлы по вертикали
for df in df_names:
    # Добавьте данные к общему DataFrame
    combined_data = pd.concat([combined_data, df], ignore_index=True)
```

```python
# Сохраните объединенные данные в новый CSV файл
combined_data.to_csv('C:/Users/NightMare/Desktop/YandexHack/DATA_well_done/combined_data_welldone.csv', index=False)
```
